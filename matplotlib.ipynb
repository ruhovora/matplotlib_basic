```python
import matplotlib.pyplot as plt
import numpy as np 

xpoints = np.array([0,6])
ypoints = np.array([0,250])

plt.plot(xpoints, ypoints)
plt.show()
```

labeling and titling


```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 ={'family':'serif', 'color': 'green' , 'size': 20}
font2 = {'family': 'lucida', 'color': 'darkblue','size': 15}
plt.xlabel("Average Pulse", fontdict=font2)
plt.ylabel("Calorie Burnage")
plt.title("Sports Watch Data", fontdict= font1, loc='left')

plt.plot(x,y)
plt.grid(axis='both', color = 'green', linestyle = "--", linewidth = 0.5)
plt.show()
```

subplot


```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.subplot(1,2,1) #1 row , 2 cloumns, and this ploit is the 1st plot
plt.plot(x,y)
plt.title("sales")
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(1,2,2) #1 row, 2 columns, and this plot is the 2nd plot
plt.plot(x,y)
plt.title("income")
```

scatter plots


```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,32,232,23,454,55,556,556,667,67,78,78,32])
plt.scatter(x,y, color = 'hotpink', s=sizes)

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y)
plt.show()
```

bars



```python
x = np.array (["A", "B", "C", "D"])
y = np.array ([3,8,1,10])

plt.bar(x,y, color = "pink", width = 0.1)
plt.show()
```

Histogram


```python
x = np.random.normal(170,10,250) #random generator, with 250 concentrated on 170 with standard diviation of 10
plt.hist(x)
```

Pie Charts


```python
y = np.array([35,25,25,15])
mylablels = ["apples", "bananas", "cherries", "dates"]
myexplode = [0.2,0.01,0,0]
plt.pie(y, labels = mylablels , explode=myexplode, shadow = True)
plt.legend(title = "Four Fruits: ")

```
